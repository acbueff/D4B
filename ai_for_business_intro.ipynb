Let's look at a practical example: Customer Segmentation using AI

```python
import numpy as np
from sklearn.cluster import KMeans

# Generate sample customer data
np.random.seed(42)
n_customers = 200

# Create customer features
customer_data = {
    'Annual_Spend': np.random.normal(1000, 500, n_customers),
    'Purchase_Frequency': np.random.normal(10, 5, n_customers)
}

# Convert to DataFrame
customer_df = pd.DataFrame(customer_data)

# Perform customer segmentation
kmeans = KMeans(n_clusters=3, random_state=42)
customer_df['Segment'] = kmeans.fit_predict(customer_df)

# Visualize segments
plt.figure(figsize=(10, 6))
sns.scatterplot(data=customer_df, x='Annual_Spend', y='Purchase_Frequency', 
                hue='Segment', palette='deep')
plt.title('Customer Segments Based on Spending Behavior')
plt.xlabel('Annual Spend ($)')
plt.ylabel('Purchase Frequency (times/year)')
plt.show()

# Analyze segments
print("\nCustomer Segment Analysis:")
print(customer_df.groupby('Segment').agg({
    'Annual_Spend': ['mean', 'count'],
    'Purchase_Frequency': 'mean'
}).round(2))
```

## Part 3: Understanding AI, Machine Learning, and Deep Learning

Let's clarify the relationship between these terms using a business context:

- **Artificial Intelligence (AI)**: The broad capability of machines to perform tasks that typically require human intelligence
- **Machine Learning (ML)**: Systems that learn from data to make decisions
- **Deep Learning (DL)**: Advanced ML using neural networks for complex pattern recognition

Let's visualize this relationship with a simple example:

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Generate sample sales data
marketing_spend = np.linspace(0, 100, 20)
sales = 2 * marketing_spend + np.random.normal(0, 10, 20)

# Reshape data for sklearn
X = marketing_spend.reshape(-1, 1)
y = sales

# Simple ML: Linear Regression
model_simple = LinearRegression()
model_simple.fit(X, y)
pred_simple = model_simple.predict(X)

# More Complex ML: Polynomial Regression
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
model_complex = LinearRegression()
model_complex.fit(X_poly, y)
pred_complex = model_complex.predict(X_poly)

# Visualize different approaches
plt.figure(figsize=(12, 6))
plt.scatter(marketing_spend, sales, color='blue', label='Actual Data')
plt.plot(marketing_spend, pred_simple, color='red', label='Simple ML (Linear)')
plt.plot(marketing_spend, pred_complex, color='green', label='Complex ML (Polynomial)')
plt.title('Marketing Spend vs. Sales: Different ML Approaches')
plt.xlabel('Marketing Spend ($1000s)')
plt.ylabel('Sales ($1000s)')
plt.legend()
plt.show()
```

## Part 4: The Intelligent Agent Paradigm

An intelligent agent is a system that:
1. Observes its environment
2. Makes decisions
3. Takes actions
4. Learns from results

Let's simulate a simple business agent that makes pricing decisions:

```python
class PricingAgent:
    def __init__(self, initial_price=100):
        self.price = initial_price
        self.sales_history = []
        self.revenue_history = []
    
    def observe_market(self, demand):
        sales = max(0, demand - self.price/10)  # Simple demand function
        revenue = sales * self.price
        self.sales_history.append(sales)
        self.revenue_history.append(revenue)
        return sales, revenue
    
    def adjust_price(self):
        if len(self.revenue_history) < 2:
            return
        
        if self.revenue_history[-1] > self.revenue_history[-2]:
            # If revenue increased, continue price direction
            self.price *= 1.1 if self.price > self.price else 0.9
        else:
            # If revenue decreased, reverse price direction
            self.price *= 0.9 if self.price > self.price else 1.1

# Simulate market conditions
agent = PricingAgent()
market_demand = np.random.normal(100, 10, 30)
prices = []
revenues = []

for demand in market_demand:
    prices.append(agent.price)
    sales, revenue = agent.observe_market(demand)
    revenues.append(revenue)
    agent.adjust_price()

# Visualize agent's performance
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

ax1.plot(prices)
ax1.set_title('Price Adjustments Over Time')
ax1.set_ylabel('Price ($)')

ax2.plot(revenues)
ax2.set_title('Revenue Over Time')
ax2.set_ylabel('Revenue ($)')
ax2.set_xlabel('Time Period')

plt.tight_layout()
plt.show()
```

## Part 5: Human-Machine Collaboration

AI works best when combined with human expertise. Let's explore a simple example of how humans and AI can work together in decision-making:

```python
from sklearn.ensemble import RandomForestClassifier

# Simulate loan application data
np.random.seed(42)
n_applications = 100

# Generate features
credit_scores = np.random.normal(700, 100, n_applications)
income = np.random.normal(60000, 20000, n_applications)
debt_ratio = np.random.normal(0.3, 0.1, n_applications)

# Create historical decisions (some by AI, some by humans)
X = np.column_stack([credit_scores, income, debt_ratio])
y = (credit_scores > 650) & (income > 40000) & (debt_ratio < 0.4)

# Train a model on historical decisions
model = RandomForestClassifier(random_state=42)
model.fit(X, y)

# Generate new applications
new_applications = pd.DataFrame({
    'Credit_Score': [680, 720, 600],
    'Income': [55000, 75000, 45000],
    'Debt_Ratio': [0.35, 0.25, 0.45]
})

# Get model predictions
predictions = model.predict_proba(new_applications)

# Display results in a business-friendly format
new_applications['AI_Approval_Confidence'] = predictions[:, 1]
new_applications['Recommendation'] = ['Review by Human' if conf > 0.3 and conf < 0.7 
                                     else 'Automatic Approval' if conf >= 0.7
                                     else 'Automatic Rejection'
                                     for conf in predictions[:, 1]]

print("Loan Application Analysis:")
print(new_applications.round(2))

# Visualize decision boundaries
plt.figure(figsize=(10, 6))
plt.scatter(credit_scores[y], income[y], c='green', label='Approved')
plt.scatter(credit_scores[~y], income[~y], c='red', label='Rejected')
plt.xlabel('Credit Score')
plt.ylabel('Income ($)')
plt.title('Loan Approval Decision Boundary')
plt.legend()
plt.show()
```
